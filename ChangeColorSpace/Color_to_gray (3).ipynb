```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path=input("Enter the path : ") 
```

```python
#Read the image
image = cv2.imread(path)
#Convert the BGR image to RGB
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#Convert the RGB image to Gray Scale image
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
#Convert the gray image to RGB
fi_gray=cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
```

```python
#Displaying of Gray iamge
cv2.imshow('Gray image', fi_gray)
plt.imshow(fi_gray)
```

```python
#Saving of image
cv2.imwrite("gray_scale.png",gray)
```
